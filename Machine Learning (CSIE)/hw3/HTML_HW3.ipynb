```python
# Q12

def get_data(filename):
    x, y = [], []
    file = open(filename, 'r')
    for line in file.readlines():
        split = line.split('\t')
        y.append(eval(split.pop()))
        x.append([1]+list(map(eval, split)))
    file.close()
    return x, y

def Q_order_transform(X, Q):
    trans_x = copy.deepcopy(X)
    for i in range(len(X)):
        temp = X[i][1:]
        for j in range(1, Q):
            trans_x[i] = trans_x[i] + [x**(j+1) for x in temp]
    return trans_x

def LinReg(X, y):
    pseudo_inverse = np.linalg.pinv(X)
    w = np.dot(pseudo_inverse, y)
    return w

def E01(X, w, y):
    l, count = len(y), 0
    Xw = np.dot(X, w)
    for i in range(l):
        count += np.sign(Xw[i]) != y[i]
    return count / l

train_x, train_y = get_data("hw3_train.dat")
test_x, test_y = get_data("hw3_test.dat")
trans_train_x, trans_test_x = Q_order_transform(train_x, 2), Q_order_transform(test_x, 2)
w = LinReg(trans_train_x, train_y)
print(abs(E01(trans_train_x, w, train_y) - E01(trans_test_x, w, test_y)))
```

```python
# Q14

def get_data(filename):
    x, y = [], []
    file = open(filename, 'r')
    for line in file.readlines():
        split = line.split('\t')
        y.append(eval(split.pop()))
        x.append([1]+list(map(eval, split)))
    file.close()
    return x, y

def full_order2_transform(X):
    trans_x = copy.deepcopy(X)
    for i in range(len(X)):
        temp = X[i][1:]
        for j in range(len(temp)):
            for k in range(j, len(temp)):
                trans_x[i] = trans_x[i] + [temp[j]*temp[k]]
    return trans_x

def LinReg(X, y):
    pseudo_inverse = np.linalg.pinv(X)
    w = np.dot(pseudo_inverse, y)
    return w

def E01(X, w, y):
    l, count = len(y), 0
    Xw = np.dot(X, w)
    for i in range(l):
        count += np.sign(Xw[i]) != y[i]
    return count / l

train_x, train_y = get_data("hw3_train.dat")
test_x, test_y = get_data("hw3_test.dat")
trans_train_x, trans_test_x = full_order2_transform(train_x), full_order2_transform(test_x)
w = LinReg(trans_train_x, train_y)
print(abs(E01(trans_train_x, w, train_y) - E01(trans_test_x, w, test_y)))
```

```python
# Q15

def get_data(filename):
    x, y = [], []
    file = open(filename, 'r')
    for line in file.readlines():
        split = line.split('\t')
        y.append(eval(split.pop()))
        x.append([1]+list(map(eval, split)))
    file.close()
    return x, y

def lower_transform(X, ith):
    trans_x = []
    for i in range(len(X)):
        trans_x.append(X[i][:ith+1])
    return trans_x

def LinReg(X, y):
    pseudo_inverse = np.linalg.pinv(X)
    w = np.dot(pseudo_inverse, y)
    return w

def E01(X, w, y):
    l, count = len(y), 0
    Xw = np.dot(X, w)
    for i in range(l):
        count += np.sign(Xw[i]) != y[i]
    return count / l

train_x, train_y = get_data("hw3_train.dat")
test_x, test_y = get_data("hw3_test.dat")
min = 999999999
index = 1
for i in range(1, 11):
    trans_train_x, trans_test_x = lower_transform(train_x, i), lower_transform(test_x, i)
    w = LinReg(trans_train_x, train_y)
    error = abs(E01(trans_train_x, w, train_y) - E01(trans_test_x, w, test_y))
    if error < min:
        min = error
        index = i

print(index)
```

```python
# Q16

def get_data(filename):
    x, y = [], []
    file = open(filename, 'r')
    for line in file.readlines():
        split = line.split('\t')
        y.append(eval(split.pop()))
        x.append([1]+list(map(eval, split)))
    file.close()
    return x, y

def random_transform(X, idxs):
    trans_x = []
    for i in range(len(X)):
        trans_x.append([1] + list(X[i][e+1] for e in idxs))
    return trans_x

def LinReg(X, y):
    pseudo_inverse = np.linalg.pinv(X)
    w = np.dot(pseudo_inverse, y)
    return w

def E01(X, w, y):
    l, count = len(y), 0
    Xw = np.dot(X, w)
    for i in range(l):
        count += np.sign(Xw[i]) != y[i]
    return count / l

train_x, train_y = get_data("hw3_train.dat")
test_x, test_y = get_data("hw3_test.dat")
times = 200
error = 0
seed = 12345
np.random.seed(seed)
for i in range(times):
    idxs = np.random.choice(np.arange(10), 5, replace = False)
    trans_train_x, trans_test_x = random_transform(train_x, idxs), random_transform(test_x, idxs)
    w = LinReg(trans_train_x, train_y)
    error += abs(E01(trans_train_x, w, train_y) - E01(trans_test_x, w, test_y))

print(error / times)
```
