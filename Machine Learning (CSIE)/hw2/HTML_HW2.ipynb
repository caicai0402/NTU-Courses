```python
# Q13

def generate_data(seed):
    np.random.seed(seed)
    N, M = 200, 5000
    train_x, train_y, test_x, test_y = [], [], [], []
    for i in range(N):
        y = np.random.choice([1, -1])
        if y == 1:
            x1, x2 = np.random.multivariate_normal([2, 3], [[0.6, 0], [0, 0.6]])
        elif y == -1:
            x1, x2 = np.random.multivariate_normal([0, 4], [[0.4, 0], [0, 0.4]])
        train_x.append([1, x1, x2])
        train_y.append(y)
    for j in range(M):
        y = np.random.choice([1, -1])
        if y == 1:
            x1, x2 = np.random.multivariate_normal([2, 3], [[0.6, 0], [0, 0.6]])
        elif y == -1:
            x1, x2 = np.random.multivariate_normal([0, 4], [[0.4, 0], [0, 0.4]])
        test_x.append([1, x1, x2])
        test_y.append(y)
    return np.array(train_x), np.array(train_y), np.array(test_x), np.array(test_y)

def LinReg(train_x, train_y):
    pseudo_inverse = np.linalg.pinv(train_x)
    w = np.dot(pseudo_inverse, train_y)
    return w

def square_length(v):
    return sum(i * i for i in v)

def E(X, y, w):
    l = len(y)
    v = np.dot(X, w) - y
    return square_length(v) / l

times = 100
ans = 0
for i in range(times):
    train_x, train_y, test_x, test_y = generate_data(i)
    w = LinReg(train_x, train_y)
    ans += E(train_x, train_y, w)
print(ans / times)
```
