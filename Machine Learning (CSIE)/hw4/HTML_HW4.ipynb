```python
import numpy as np
import copy
import math
from sklearn.preprocessing import PolynomialFeatures
```

```python
def get_data_and_transform(filename, order = 1):
    x, y = [], []
    file = open(filename, 'r')
    for line in file.readlines():
        split = line.split(' ')
        y.append(eval(split.pop()))
        x.append(list(map(eval, split)))
    file.close()
    poly = PolynomialFeatures(order)
    x_poly = poly.fit_transform(x)
    return x_poly, y

train_x, train_y = get_data_and_transform("hw4_train.dat", 3)
test_x, test_y = get_data_and_transform("hw4_test.dat", 3)
```

```python
# Q16

v_fold = 5
lamb = [-4, -2, 0, 2, 4]
best_acc = 0

for l in lamb:
    temp = 0
    param = parameter(f'-s 0 -c {1/(2*10**l)} -e 0.000001 -q')
    for i in range(v_fold):
        size = int(len(train_x) / v_fold)
        v_train_x, v_train_y, v_test_x, v_test_y = [], [], [], []
        for j in range(v_fold):
            if j != i:
                v_train_x = v_train_x + list(train_x[size * j : size * (j+1)])
                v_train_y = v_train_y + list(train_y[size * j : size * (j+1)])
            else:
                v_test_x = v_test_x + list(train_x[size * j : size * (j+1)])
                v_test_y = v_test_y + list(train_y[size * j : size * (j+1)])
        prob  = problem(v_train_y, v_train_x)
        model = train(prob, param)
        p_label, p_acc, p_val = predict(v_test_y, v_test_x, model)
        temp = temp + p_acc[0]
    if temp / v_fold > best_acc:
        best_acc = temp / v_fold

print(1 - best_acc/100)
```
