```python
import numpy as np
!pip install -U libsvm-official
from libsvm.svmutil import *
import copy
from IPython.display import clear_output
```

```python
# Q11

train_y, train_x = svm_read_problem('satimage.scale', return_scipy = True)
train_x = train_x.toarray()
dim = len(train_x[0])

train_y_5 = copy.deepcopy(train_y)
for i in range(len(train_y_5)):
    train_y_5[i] = 2 * (train_y_5[i] == 5) - 1

prob = svm_problem(train_y_5, train_x)
param = svm_parameter('-s 0 -t 0 -c 10 -e 0.000001')
model = svm_train(prob, param)
sv = np.asarray(model.get_SV())
sv_coef = np.asarray(model.get_sv_coef())
sv_indices = np.asarray(model.get_sv_indices())

w = np.zeros(dim)
for index, coef in zip(sv_indices, sv_coef):
    index = index - 1
    alpha = coef[0]
    for row in range(dim):
        w[row] = w[row] + alpha * train_x[index][row]
        
print(np.linalg.norm(w))
```

```python
# Q12

train_y, train_x = svm_read_problem('satimage.scale', return_scipy = True)
train_x = train_x.toarray()
dim = len(train_x[0])
labels = [2, 3, 4, 5, 6]
label = 0
acc = 100

for l in labels:
    train_y_l = copy.deepcopy(train_y)
    for i in range(len(train_y_l)):
        train_y_l[i] = 2 * (train_y_l[i] == l) - 1
    prob = svm_problem(train_y_l, train_x)
    param = svm_parameter('-s 0 -t 1 -c 10 -d 3 -e 0.000001 -r 1 -g 1')
    model = svm_train(prob, param)
    p_label, p_acc, p_val = svm_predict(train_y_l, train_x, model)
    if p_acc[0] < acc:
        label = l
        acc = p_acc[0]

print(label)
```

```python
# Q14

train_y, train_x = svm_read_problem('satimage.scale', return_scipy = True)
test_y, test_x = svm_read_problem('satimage.scale.t', return_scipy = True)
train_x = train_x.toarray()
test_x = test_x.toarray()
dim = len(train_x[0])

train_y_1 = copy.deepcopy(train_y)
test_y_1 = copy.deepcopy(test_y)
for i in range(len(train_y_1)):
    train_y_1[i] = 2 * (train_y_1[i] == 1) - 1
for i in range(len(test_y_1)):
    test_y_1[i] = 2 * (test_y_1[i] == 1) - 1

Cs = [0.01, 0.1, 1, 10, 100]
prob = svm_problem(train_y_1, train_x)
best_C = 0.01
acc = 0

for C in Cs:
    param = svm_parameter(f'-s 0 -t 2 -c {C} -g 10 -e 0.000001')
    model = svm_train(prob, param)
    p_label, p_acc, p_val = svm_predict(test_y_1, test_x, model)
    if p_acc[0] > acc:
        best_C = C
        acc = p_acc[0]
 
print(best_C)
```

```python
# Q16

train_y, train_x = svm_read_problem('satimage.scale', return_scipy = True)
train_x = train_x.toarray()
dim = len(train_x[0])
train_y_1 = copy.deepcopy(train_y)
for i in range(len(train_y_1)):
    train_y_1[i] = 2 * (train_y_1[i] == 1) - 1

rs = [0.1, 1, 10, 100, 1000]
count = [0, 0, 0, 0, 0]
result = []
times = 1000

for i in range(times):
    print(i, count)
    temp = list(zip(train_y_1, train_x))
    np.random.shuffle(temp)
    test_y_val, test_x_val = zip(*temp[:200])
    train_y_val, train_x_val = zip(*temp[200:])
    prob = svm_problem(train_y_val, train_x_val)
    best_r, acc = 0, 0
    for r in range(len(rs)):
        param = svm_parameter(f'-s 0 -t 2 -c 0.1 -g {rs[r]} -e 0.000001 -q')
        model = svm_train(prob, param)
        p_label, p_acc, p_val = svm_predict(test_y_val, test_x_val, model)
        if p_acc[0] > acc:
            best_r = r
            acc = p_acc[0]
    clear_output()
    count[best_r] += 1
    if times / 10 == 0:
        result.append(count)

print(rs[count.index(max(count))])
```
