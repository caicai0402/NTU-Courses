```python
import numpy as np
import random
```

```python
# Q11

def f(x1, x2):
    return np.sign(x1)

def h1(x1, x2):
    return np.sign(2 * x1 - x2)

def h2(x1, x2):
    return np.sign(x2)

ans = 0
times = 1000000

for i in range(times):
    count1 = 0
    count2 = 0
    for j in range(4):
        x1 = 2 * random.random() - 1
        x2 = 2 * random.random() - 1
        f_result = f(x1, x2)
        h1_result = h1(x1, x2)
        h2_result = h2(x1, x2)
        if f_result != h1_result:
            count1 += 1
        if f_result != h2_result:
            count2 += 1
    if count1 == count2:
        ans += 1

print(ans / times * 4096)
```

```python
# Q12

Dices = [[{2, 4, 6}, {1, 3, 5}],
         [{1, 2, 6}, {3, 4, 5}],
         [{6}, {1, 2, 3, 4, 5}],
         [{2, 3, 5}, {1, 4, 6}]]
ans = 0
for a in range(4):
    for b in range(4):
        for c in range(4):
            for d in range(4):
                for e in range(4):
                    ans += len(Dices[a][0].intersection(Dices[b][0], Dices[c][0], Dices[d][0], Dices[e][0])) > 0

print(ans)
```

```python
# Q13

def square_length(x):                              # 計算 vector 長度
    return sum(i * i for i in x)

def get_data(filename):                            # 讀檔並做資料前處理
    data = []
    file = open(filename, 'r')
    for line in file.readlines():
        split = line.split('\t')
        y = eval(split.pop())
        x = [1]+list(map(eval, split))
        data.append((x, y))
    file.close()
    return data
  
def my_sign(num):                                  # 定義一個 sign(0) = -1 的的 sign
    if num > 0:
        return 1
    else:
        return -1

def check(w, x, y):                                # 檢查 weight 正確或錯誤
    if my_sign(np.inner(w, x)) == y :
        return True
    else:
        return False

def train(N, data, seed):                          # 定義訓練行為
    random.seed(seed)
    w = [0 for i in range(11)]
    count = 0
    while count < 5*N:
        index = random.randrange(0, 100)
        x = data[index][0]
        y = data[index][1]
        if check(w, x, y) == True:
            count += 1
        else:
            count = 0
            w = list(np.array(w) + y * np.array(x))
    return w

filename = "hw1_train.dat.txt"                     # 設定變數
data = get_data(filename)
N = 100
times = 1000
w_square = 0

for seed in range(times):                          # 開 train
    w = train(N, data, seed+123456)
    w_square += square_length(w)

print(w_square / times)                            # 輸出答案
```
